```python
# Install necessary libraries (if not already installed)
!pip install transformers
!pip install torch
!pip install datasets
!pip install pandas

import pandas as pd
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load the dataset from the local CSV file
df = pd.read_csv('/content/Customer-Support.csv')

# Extract query and response data from the dataset
queries = df['query'].tolist()
responses = df['response'].tolist()

# Load the pre-trained GPT-2 model and tokenizer
model = GPT2LMHeadModel.from_pretrained('gpt2')
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
tokenizer.add_special_tokens({'pad_token': '[PAD]'})  # Set pad_token

# Function to preprocess and tokenize the dataset
def preprocess_dataset(queries):
    tokenized_queries = tokenizer(queries, truncation=True, padding="max_length", return_tensors="pt")
    return tokenized_queries

# Preprocess and tokenize the dataset
tokenized_queries = preprocess_dataset(queries)

# Define a function to generate responses
def generate_response(input_query):
    input_ids = tokenizer.encode(input_query, return_tensors='pt')
    max_length = 100
    response = model.generate(input_ids, max_length=max_length, num_return_sequences=1, pad_token_id=tokenizer.pad_token_id)
    return tokenizer.decode(response[0], skip_special_tokens=True)

# Sample query for testing
query = "I have an issue with my order delivery."
response = generate_response(query)
print("Automated Response:", response)
```
